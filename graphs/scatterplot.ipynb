# Scatterplot chart

```python
import matplotlib.pyplot as plt
```

```python
x = [ 1, 2, 5 ]
y = [ 1, 3, 7 ]
title = 'Scatterplot: gráfico de dispersão'
axis_x = 'X axis'
axis_y = 'Y axis'
```

```python
plt.title(title)

plt.xlabel(axis_x)
plt.ylabel(axis_y)

plt.scatter(x, y)
plt.plot(x, y)
plt.show()
```
